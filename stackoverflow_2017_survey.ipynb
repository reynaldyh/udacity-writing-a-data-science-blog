```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualization
```

## Import Dataset

```python
data_collections = dict()
schema_collections = dict()

for year in range(2017, 2021):
    data_collections[str(year)] = pd.read_csv(f"dataset/{year}/survey_results_public.csv")
    schema_collections[str(year)] = pd.read_csv(f"dataset/{year}/survey_results_schema.csv")
```

## Question 1: The trend of the number of respondents that filling the survey?

```python
respondent_trend = []
for year_date in data_collections.keys():
    respondent_trend.append(len(data_collections[year_date]))

fig, ax = plt.subplots(1, 1, figsize=(9, 7), sharex=True, sharey=True)
plt.ticklabel_format(style='plain', axis='x', useOffset=False)
plt.bar(np.arange(2017,2021), respondent_trend)
ax.set_xlabel('year')  # Add an x-label to the axes.
ax.set_ylabel('total respondents')  # Add a y-label to the axes.
ax.set_title("Trend of the respodent over 2017-2020")  # Add a title to the axes.
plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(1))
```

## Data Preparation

```python
salary_mapper = {
    '2017': 'Salary',
    '2018': 'ConvertedSalary',
    '2019': 'ConvertedComp',
    '2020': 'ConvertedComp',
}

hobyist_mapper = {
    '2017': 'ProgramHobby',
    '2018': 'Hobby',
    '2019': 'Hobbyist',
    '2020': 'Hobbyist',
}

df_hobbyist = pd.DataFrame({})
df_salary = pd.DataFrame({})

for year in data_collections.keys():
    data_collections[year]['Year'] = int(year)
    data_collections[year].rename(columns={salary_mapper[year]: 'AnnualizedConvSalary'}, inplace=True)
    data_collections[year].rename(columns={hobyist_mapper[year]: 'IsProgrammingHobby'}, inplace=True)
    df_salary = pd.concat([df_salary, data_collections[year][['AnnualizedConvSalary','Year']]])
    df_hobbyist = pd.concat([df_hobbyist, data_collections[year][['IsProgrammingHobby','Year']]])
```

#### Check Null

```python
print('Null data per columns\n',df_salary.isnull().mean())
```

```python
df_salary.dropna(inplace=True)
```

```python
print('Null data per columns\n', df_salary.isnull().mean())
```

## Question 2: The trend of average salary of the Stack Overflow’s respondent over 2017–2020?

```python
import matplotlib.ticker as mticker

fig, ax = plt.subplots(1, 1, figsize=(9, 7), sharex=True, sharey=True)
plt.bar(np.arange(2017,2021), df_salary.groupby('Year').mean()['AnnualizedConvSalary'].values)
plt.ticklabel_format(style='plain', axis='x', useOffset=False)
ax.set_xlabel('Year')  # Add an x-label to the axes.
ax.set_ylabel('Average Salary in USD')  # Add a y-label to the axes.
ax.set_title("Trend of average salary over 2017-2020")  # Add a title to the axes.
plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(1))
```

```python
df_hobbyist['IsProgrammingHobby'].unique()
```

As you can see, we need to preprocessed the data, from the pattern the positive one always starts with "Yes" pattern, so we can use startswith method, but before that we need to remove nan value first.

```python
df_hobbyist.dropna(subset=['IsProgrammingHobby'], inplace=True)
```

```python
df_hobbyist['IsProgrammingHobby'] = df_hobbyist['IsProgrammingHobby'].str.startswith('Yes')
```

## Question 3: The trend of programming as a hobby over 2017–2020?

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 7), sharex=True, sharey=True)
plt.ticklabel_format(style='plain', axis='x', useOffset=False)
plt.bar(np.arange(2017,2021), df_hobbyist.groupby('Year').mean()['IsProgrammingHobby'].values*100)
ax.set_xlabel('Year')  # Add an x-label to the axes.
ax.set_ylabel('Programming as Hobby (%)')  # Add a y-label to the axes.
ax.set_title("Trend of programming as hobby 2017-2020")  # Add a title to the axes.
plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(1))
```

Conclusion:<br>
1. Number of respondent significantly on a downtrend from 2018.<br>
2. The average salary on a uptrend until 2019.<br>
3. Programming as a hobby has slightly downtrend over the years.<br>
